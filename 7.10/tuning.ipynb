```python
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
import seaborn as sns
from scipy.stats import gaussian_kde
```

```python
def estimate_maxima(data):

    kde = gaussian_kde(data)

    samples = np.arange(min(data), max(data), 1e-9)

    probs = kde.evaluate(samples)
    maxima_index = probs.argmax()

    maxima = samples[maxima_index]

    return maxima
```

```python
%matplotlib notebook
names = []

V = []
for name in listdir():
    
    if name[-3:]=='csv':
        #print(name)
        data = np.loadtxt(name, skiprows=15, delimiter=';',
                          usecols=(3), dtype='str')

        for i in range(len(data)):
            if data[i]==' No device found':
                data[i] = 0
            else:
                data[i] = data[i].replace(',', '.')
        #print(data)
        data = data.astype(np.float32)
        
        sns.displot(data, label=name, kde=True)
        
        V.append(estimate_maxima(data))
        
        names.append(int(name[:name.index('V')]))
        #fig = sns_plot.get_figure()
        #plt.plot(data, linewidth=1, label=name)
        plt.legend()
        plt.xlabel('V, Volt')
plt.show()
```

```python
plt.title('W(V) red')
plt.xlabel('V, Volt')
plt.ylabel('W, Watt')
plt.plot(names, V, '.')
```

```python
from scipy.optimize import curve_fit

def f(U, d_U, d_I, alpha):    
    #alpha = 0.3e3 #
    x = (U-d_U)/(2*alpha)
    return np.sin(x)**2 + d_I

def g(U, d_U, alpha, d_I):
    return ((U-d_U)/(2*alpha))**2+d_I

I0 = 3e-3
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(9,4), sharey=True)

W = V_r/I0
U = names

ax1.plot(U, W, label='data_red', color='tomato')

popt, pcov = curve_fit(f, U, W, p0=[0,0, 300])
ax1.plot(U, f(U, *popt), 'g--', label=r'$fit: \alpha=%5.1f $' % popt[2])
ax1.grid()
ax1.legend()
W = V_b/I0
U = names

ax2.plot(U, W, label='data_blue')

popt, pcov = curve_fit(f, U, W, p0=[0,0, 300])
ax2.plot(U, f(U, *popt), 'g--', label=r'$fit: \alpha=%5.1f $' % popt[2])
ax2.legend()
ax2.grid()

W = V_a/I0
U = names_a

plt.plot(U, W, label='data_both')

popt, pcov = curve_fit(f, U, W, p0=[0,0, 300])
ax3.plot(U, f(U, *popt), 'g--', label=r'$fit: \alpha=%5.1f $' % popt[2])
ax3.legend()
ax3.grid()
```
